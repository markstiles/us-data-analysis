```python
def print_info(variable, df, df2, ascend, view_filter_cond=None):
    
    df1 = df[~pd.isna(place_df[variable])]
    
    view_df = df1 if view_filter_cond is None else df1[view_filter_cond]
    
    plt.hist(view_df[variable], edgecolor='black', rwidth=0.75, bins=40)
    title = variable.replace("_", " ")
    plt.title(f"{title}")
    plt.axvline(df2[variable], color='red', linestyle='dashed', linewidth=1)
    plt.show()

    print(f"{variable} range from {df1[variable].min()} to {df1[variable].max()}")
    count_less_than_waltham = len(df1[df1[variable] < df2[variable]])
    print(f"num less than waltham: {count_less_than_waltham}")
    count_total = len(df1)
    print(f"num total: {count_total}")
    percentile = count_less_than_waltham / count_total * 100
    print(f"Waltham: {df2[variable]}")
    print(f"{percentile} Percentile")
    print(df1[["Place_Name", variable]].sort_values(by=[variable], ascending=ascend)[0:10])
```

```python
# for each point of industry count, try to get the percentage of each industry 
# it seems like education, healthcare, arts and transportation are the bellweather for large economies
nums = range(1,place_df.Industry_Count.max()+1)
temp_df = pd.DataFrame()
for i in nums:
    rev_df = place_df[place_df.Industry_Count == i][[
        'Waste_Management_Revenue', 'Food_Services_Revenue', 'Retail_Revenue', 'Technical_Revenue', 'Healthcare_Revenue', 
        'Arts_Revenue', 'Other_Revenue', 'Transportation_Revenue', 'Education_Revenue'
    ]]
    ind_sums = (rev_df > 0).astype(int).sum(axis=0)
    temp_sums = round((ind_sums / ind_sums.sum()) * 100, 0)
    #print(temp_sums)
    trans = temp_sums.to_frame().T
    trans["Industry_Count"] = i
    pop_array = place_df[place_df.Industry_Count == i]['Population']
    trans["Average_Population"] = round(pop_array.sum() / pop_array.size, 0)
    temp_df = pd.concat([temp_df, trans])
    
temp_df = temp_df.fillna(0)
temp_df = temp_df.set_index('Industry_Count')
temp_df
```
