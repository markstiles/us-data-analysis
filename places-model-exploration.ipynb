```python
# for each point of industry count, try to get the percentage of each industry 
# it seems like education, healthcare, arts and transportation are the bellweather for large economies
nums = range(1,place_df.Industry_Count.max()+1)
temp_df = pd.DataFrame()
for i in nums:
    rev_df = place_df[place_df.Industry_Count == i][[
        'Waste_Management_Revenue', 'Food_Services_Revenue', 'Retail_Revenue', 'Technical_Revenue', 'Healthcare_Revenue', 
        'Arts_Revenue', 'Other_Revenue', 'Transportation_Revenue', 'Education_Revenue'
    ]]
    ind_sums = (rev_df > 0).astype(int).sum(axis=0)
    temp_sums = round((ind_sums / ind_sums.sum()) * 100, 0)
    #print(temp_sums)
    trans = temp_sums.to_frame().T
    trans["Industry_Count"] = i
    pop_array = place_df[place_df.Industry_Count == i]['Population']
    trans["Average_Population"] = round(pop_array.sum() / pop_array.size, 0)
    temp_df = pd.concat([temp_df, trans])
    
temp_df = temp_df.fillna(0)
temp_df = temp_df.set_index('Industry_Count')
temp_df
```
